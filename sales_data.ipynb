Create a DataFrame

```python
import pandas as pd
data = { 
    "Category" : ["Pen", "Pencil", "Notebooks", "Bags", "Pouch"] ,
    "Sales" : [55000, 75000, 65000, 25000, 50000]
    }
df = pd.DataFrame(data)

print("Yearly Stationary sales report:")
print(df)


```

Aggregate Data

```python
category_sales = df.groupby('Category')['Sales'].sum()
print(category_sales)
```

Bar Chart

```python
import matplotlib.pyplot as plt
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C299FF']
plt.bar(category_sales.index, category_sales.values, color=colors)
plt.xlabel = "Categories"
plt.ylabel = "Total Sales"
plt.title("Yearly Stationery Sales Report")
plt.show()

print(df)
```

Pie chart

```python
plt.figure(figsize=(6,6))
plt.pie(
    category_sales.values,               
    labels=category_sales.index,          
    autopct='%1.1f%%',                    
    startangle=90,                        
    colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C299FF']                            
)

plt.title("Sales Distribution by Category")
plt.show()
```

Subplots

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# --- Bar Chart ---
axes[0].bar(category_sales.index, category_sales.values, 
            color=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C299FF'])
axes[0].set_xlabel("Categories")
axes[0].set_ylabel("Total Sales")
axes[0].set_title("Bar Chart: Sales by Category")


# --- Pie Chart ---
axes[1].pie(
    category_sales.values,
    labels=category_sales.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C299FF'],
)
axes[1].set_title("Pie Chart: Sales Distribution")

plt.tight_layout()
plt.show()
```
