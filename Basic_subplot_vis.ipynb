Basic Subplot Visualization

```python
import pandas as pd
data = {
    "Salary" : [40000, 70000, 50000, 60000, 30000],
    "Bonus" : [0.5, 0.7, 0.4, 0.6, 0.9]
}

df = pd.DataFrame(data)
print("Bonus calculation")
print(df)

```

Aggregrate data

```python
Bonus = df.groupby('Salary')['Bonus'].sum()
print(df)
```

Line chart

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
plt.plot(Bonus.index, Bonus.values, marker='*', linestyle='-', color="magenta")
plt.show()
```

Scatter plot

```python
plt.figure(figsize=(4,4))
plt.scatter(Bonus.index, Bonus.values, marker='o', color="teal")
plt.show()
```

Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# line chart
axes[0].plot(Bonus.index, Bonus.values, marker='*', linestyle='-', color="magenta")
axes[0].set_xlabel("Salary")
axes[0].set_ylabel("Bonus")
axes[0].set_title("Line chart: Bonus Calculation")

axes[1].scatter(Bonus.index, Bonus.values, marker='o', color="teal")
axes[1].set_xlabel("Salary")
axes[1].set_ylabel("Bonus")
axes[1].set_title("scatter plot: Bonus variation")


plt.show()




# scatter plot
```
