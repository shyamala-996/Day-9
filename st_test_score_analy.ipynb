Student Test Score Analysis

Load data:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    "School": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "Score": [88, 92, 79, 85, 90, 87, 91, 95, 89]
})
```

Score distribution (Histogram):

```python
plt.figure(figsize=(6,4))
sns.histplot(data["Score"], bins=5, color = "teal", kde=True)
plt.title("Overall Test Score Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

Compare the school scores (Boxplot):

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="School", y="Score", color="magenta", data=data)
plt.title("Test Scores by School")
plt.show()
```
